**1. Write a recursive program to solve The Tower of Hanoi Problem.**

```
def towerOfHanoi(n: int, source: str, auxilliary: str, destination: str):
    
    if n <= 0:
        return
    
    towerOfHanoi(n - 1, source, destination, auxilliary)
    print("Moved disk", n, "from", source, "to", destination) 
    towerOfHanoi(n - 1, auxilliary, source, destination)
    return

if __name__ == "__main__": 
    n = int(input())
    towerOfHanoi(n, "A", "B", "C")
```

**2. Write a recursive function to calculate the sum of digits of a n digit number.**

```
def sumOfDigits(n: int) -> int:
    if n < 10:
        return n
    return n % 10 + sumOfDigits(n // 10)
 
 
if __name__ == "__main__":
    number = int(input())
    print("The sum of digits of the number entered is: ", sumOfDigits(number))
```

**3. Write a recursive function to find the binary equivalent of a positive integer.**

```
def intToBinary(n: int) -> str:
    if n == 0:
        return ""
    return intToBinary(n // 2) + str(n % 2)

if __name__ == "__main__":
    integer = int(input())
    print("The Binary representation of the inputed number is: ", intToBinary(integer))
```

**4. Write a recursive program without Tail recursion to calculate sum of first n natural numbers.**

```
def sumOfNaturalNumbers(n: int) -> int:
    if n == 1:
        return 1
    return n + sumOfNaturalNumbers(n - 1)

if __name__ == "__main__":
    n = int(input())
    print("The sum of all natural numbers till n is: ", sumOfNaturalNumbers(n))
```

**5. Write the function which removes all the vowels in a word and returns the remaining letters in the word.**

```
def removeVowels(word: str) -> str:
    return "".join([letter for letter in word if letter not in "AEIOUaeiou"])

if __name__ == "__main__":
    word = input()
    print("The word after removing all the vowels is: ", removeVowels(word))
```

**6. Write a program to convert a hexadecimal number entered as a string into its equivalent binary number.**

```
def hexToBinary(hexadecimal: str) -> str:
    binary_equivalent = {'0': '0000', '1': '0001', '2': '0010', '3': '0011', '4': '0100', '5': '0101', '6': '0110', '7': '0111', '8': '1000', '9': '1001', 'A': '1010', 'B': '1011', 'C': '1100', 'D': '1101', 'E': '1110', 'F': '1111'}
    return ("".join([binary_equivalent[each] for each in hexadecimal])).lstrip('0')

if __name__ == "__main__":
    hexadecimal = input()
    print("The binary equivalent of the given hexadecimal number is: ", hexToBinary(hexadecimal))
```

```
def linearSearch(nums: [int], n: int) -> bool:
    for each in nums:
        if each == n:
            return True
    return False

if __name__ == "__main__":
    nums = list(map(int, input().split()))
    n = int(input())
    if linearSearch(nums, n):
        print("The number is present in the list")
    else:
        print("The number is not present in the list")
```

**9. Write a program to implement binary search.**

```
def binarySearch(nums: [int], n: int) -> bool:
    sorted(nums)
    left = 0
    right = len(nums) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if nums[mid] == n:
            return True
        if n > nums[mid]:
            left = mid + 1
        if n < nums[mid]:
            right = mid - 1
    return False
if __name__ == "__main__":
    nums = list(map(int, input().split()))
    n = int(input())
    if binarySearch(nums, n):
        print("The number is present in the list")
    else:
        print("The number is not present in the list")
```

**10. Write a program to implement insertion sort.**

```
def insertionSort(numbers: [int]) -> [int]:
    for i in range(1, len(numbers)):
        index = i
        while index > 0 and numbers[index] < numbers[index - 1]:
            numbers[index], numbers[index - 1] = numbers[index - 1], numbers[index]
            index = index - 1
    return numbers

if __name__ == "__main__":
    numbers = list(map(int, input().split()))
    numbers = insertionSort(numbers)
    print("The numbers in sorted order is: ")
    print(numbers)
```

**11. Write a program to implement merge sort.**

```
def merge(array1: [int], array2: [int]) -> [int]: 
    index1 = 0;
    index2 = 0;
    sorted_array = []
    while index1 < len(array1) and index2 < len(array2):
        if array1[index1] < array2[index2]:
            sorted_array.append(array1[index1])
            index1 += 1
        elif array2[index2] < array1[index1]:
            sorted_array.append(array2[index2])
            index2 += 1
        elif array1[index1] == array2[index2]:
            sorted_array.append(array1[index1])
            sorted_array.append(array2[index2])
            index1 += 1
            index2 += 1
    while index1 < len(array1):
        sorted_array.append(array1[index1])
        index1 += 1
    while index2 < len(array2):
        sorted_array.append(array2[index2])
        index2 += 1
    return sorted_array

def mergeSort(array: [int]) -> [int]:

    if len(array) <= 1:
        return array
    middle_index = len(array) // 2
    left_half = array[: middle_index]
    right_half = array[middle_index: ]
    return merge(mergeSort(left_half), mergeSort(right_half))

if __name__ == "__main__":
    numbers = list(map(int, input().split()))
    numbers = mergeSort(numbers)
    print("The numbers after sorting: ")
    print(numbers)
```

**12. Write a program to implement quick sort.**

```
def quickSort(array: [int], starting_index: int, ending_index: int) -> [int]:
    if starting_index >= ending_index:
        return array
    pivot_index = starting_index
    left_index = starting_index + 1
    right_index = ending_index
    while right_index >= left_index:
        if array[left_index] > array[pivot_index] and array[right_index] < array[pivot_index]:
            array[left_index], array[right_index] = array[right_index], array[left_index]
        elif array[left_index] <= array[pivot_index]:
            left_index += 1
        elif array[right_index] >= array[pivot_index]:
            right_index -= 1
    array[pivot_index], array[right_index] = array[right_index], array[pivot_index]
    isLeftSmaller = right_index - 1 - starting_index < ending_index - (right_index + 1)
    if isLeftSmaller:
        quickSort(array, starting_index, right_index - 1)
        quickSort(array, right_index + 1, ending_index)
    else:
        quickSort(array, right_index + 1, ending_index)
        quickSort(array, starting_index, right_index -1)
    return array

if __name__ == "__main__":
    numbers = list(map(int, input().split()))
    numbers = quickSort(numbers, 0, len(numbers) - 1)
    print("The numbers after sorting is: ")
    print(numbers)
```

**13. Write a program to perform the following operations on complex numbers using class.**

    a) Addition  
    b) Subtraction  
    c) Multiplication  
    d) Check two complex numbers are equal or not

```
class Complex:

    def __init__(self, x: int, y: int):
        self.real = x
        self.imaginary = y 

    def add(self, number):
        return Complex((self.real + number.real), (self.imaginary + number.imaginary))

    def subtract(self, number):
        return Complex((self.real - number.real), (self.imaginary - number.imaginary))

    def multiply(self, number):
        real = (self.real * number.real) - (self.imaginary * number.imaginary)
        imaginary = (self.real * number.imaginary) + (number.real * self.imaginary)
        return Complex(real, imaginary)

    def isEqual(self, number) -> bool:
        isRealEqual = self.real == number.real
        isImaginaryEqual = self.imaginary == number.imaginary
        return isRealEqual and isImaginaryEqual

    def show(self):
        if self.imaginary < 0:
            print("%d - %di"%(self.real, abs(self.imaginary)))
            return
        print("%d + %di"%(self.real, self.imaginary))
        return

if __name__ == "__main__":
    c1 = Complex(2, 3)
    c2 = Complex(4, -7)
    c3 = c1.add(c2)
    c4 = c1.subtract(c2)
    c5 = c1.multiply(c2)
    c6 = Complex(29, -2)
    c1.show()
    c2.show()
    c3.show()
    c4.show()
    c5.show()
    isEqual1 = c5.isEqual(c6)
    isEqual2 = c3.isEqual(c4)
    print(isEqual1)
    print(isEqual2)
```

**14. Write a program to copy the contents of one file to another.**

```
def copyContent(copy_from: str, copy_to: str):
    file1 = open(copy_from, "r")
    file2 = open(copy_to, "w")
    for each_line in file1:
        file2.write(each_line)
    file1.close()
    file2.close()
    return

def showFileContent(file_name: str):
    file1 = open(file_name, "r")
    for each_line in file1:
        print(each_line)
    return

if __name__ == "__main__":
    copyContent("file1.txt", "file2.txt")
    print("Content in First File: ")
    showFileContent("file1.txt")
    print("Content in Second File: ")
    showFileContent("file2.txt")
```
