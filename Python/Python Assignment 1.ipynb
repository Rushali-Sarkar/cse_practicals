``5. Given 3 points (x1, y1), (x2, y2) and (x3, y3) Write a program to check if all the three points fall on the straight line.``

```python
def areThreePointsinLine(point1: [int, int], point2: [int, int], point3: [int, int]) -> bool:
    x1, y1 = point1
    x2, y2 = point2
    x3, y3 = point3
    
    m = (y2 - y1) / (x2 - x1)
    c = y2 - (m * x2)
    
    return y3 == (m * x3) + c

point1 = list(map(int, input().split(",")))
point2 = list(map(int, input().split(",")))
point3 = list(map(int, input().split(",")))

if (areThreePointsinLine(point1, point2, point3)):
    print("The three points are in line")
else:
    print("The three points are not in line")
```

``6. Given the coordinates (x, y) of center of a circle and its radius, write a program that will determine whether a point lies inside the circle, on the circle or outside the circle. ``

```python
def positionofPoint(radius: int, center: [int, int], point: [int, int]) -> int:
    
    h, k = center
    x, y = point
    
    return ((x - h) ** 2 + (y - k) ** 2) - (radius * radius)

radius = int(input())
center = list(map(int, input().split(",")))
point = list(map(int, input().split(",")))

position = positionofPoint(radius, center, point)

if position == 0:
    print("The point lies on the circle")
elif position < 0:
    print("The point lies inside the circle")
elif position > 0:
    print("The point lies outside the circle")
```

``7. Write a program to print all prime numbers from 1 to 300.``

```python
def isPrime(number: int) -> bool:
    
    if number == 0 or number == 1:
        return False

    if number == 2:
        return True
    
    square_root = int(number ** (1 / 2)) + 1
    for i in range(2, square_root):
        if number % i == 0:
            return False
    return True

PRIMES = []
for number in range(1, 301):
    
    if isPrime(number):
        PRIMES.append(number)

print(PRIMES)
```

``8. Write a program to print all Armstrong number between 1 and 500.``

```python
def numberOfDigits(number: int) -> int:
    return len(str(number))

def getDigits(number: int) -> [int]:
    return [int(each) for each in str(number)]

def sumOfRaisedDigits(digits: [int], power: int) -> int:
    return sum([each ** power for each in digits])

def isArmstrong(number: int) -> bool:
    return number == sumOfRaisedDigits(getDigits(number), numberOfDigits(number))

for number in range(1, 501):
    if (isArmstrong(number)):
        print(number)
```

``10. Write a program to implement a Stack Data Structure.``

```python
class Stack:
    
    def __init__(self, capacity):
        
        self.capacity = capacity
        self.top = -1
        self.stack_arr = [0 for i in range(capacity)]

    def isEmpty(self) -> bool:
        return self.top == -1
    
    def getTop(self) -> int:
        
        if self.isEmpty():
            print("Stack Underflow")
            return -1
        return self.stack_arr[self.top]
        
    def push(self, item: int) :
        
        if self.top == self.capacity - 1:
            print("Stack Overflow")
            return
        
        self.top = self.top + 1
        self.stack_arr[self.top] = item
        return
    
    def pop(self) -> int:
        
        if isEmpty():
            print("Stack Underflow")
            return -1
        
        item = self.stack_arr[self.top]
        self.top = self.top - 1
        return item
```

``11. Write a program to implement a Queue Data Structure.``

```python
class Queue:
    
    def __init__(self, capacity):
        
        self.front = -1
        self.back = 0
        self.queue_arr = [0 for i in range(capacity)]
        self.capacity = capacity
        self.length = 0
        
    def isEmpty(self) -> bool:
        return self.length == 0
    
    def getFront(self) -> int:
        
        if self.length == 0:
            print("Queue Underflow")
            return -1
        
        return self.queue_arr[self.front]
    
    def getBack(self) -> int:
        
        if self.length == 0:
            print("Queue Underflow")
            return -1
        
        return self.queue_arr[self.back]
    
    def enqueue(self, item: int):
        
        if self.length == self.capacity:
            print("Queue Overflow")
            return
        
        if self.front == self.capacity - 1:
            self.front = -1
            
        self.front = self.front + 1
        self.queue_arr[self.front] = item
        self.length = self.length + 1
        return
    
    def dequeue(self) -> int:
        
        if self.length == 0:
            print("Queue Underflow")
            return -1
        
        if self.back == self.capacity:
            self.back = 0
            
        item = self.queue_arr[self.back]
        self.back = self.back + 1
        self.length = self.length - 1
        return item
        
```

``12. Suppose a list has 15 numbers. Write a program that removes all the duplicates from the list.``

```python
def removeDuplicates(numbers: [int]) -> [int]:
    
    trackDuplicate = {}
    new_list = []
    
    for each in numbers:
        
        if each not in trackDuplicate:
            trackDuplicate[each] = True
            new_list.append(each)
            
    return new_list

numbers = list(map(int, input().split()))
numbers = removeDuplicates(numbers)
print(numbers)
```

``13. Suppose a date is represented as a tuple (d, m, y). Write a program to create two date tuples and find the number of days between the two dates. ``

```python
import random

DAYS_IN_MONTHS = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def isLeapYear(year: int) -> bool:
    
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            return False
        return True
    return False
                
                
def updateFebruary(year: int):
    
    if isLeapYear(year):
        DAYS_IN_MONTHS[2] = 29
        
    return

def createRandomDate() -> (int):
    year = random.randrange(2000, 2051)
    updateFebruary(year)
    month = random.randrange(1, 13)
    day = random.randrange(1, DAYS_IN_MONTHS[month] + 1)
    return day, month, year


def returnEarlierDate(date1: (int), date2: (int)) -> (int):
    
    d1, m1, y1 = date1
    d2, m2, y2 = date2
    
    if y1  < y2:
        return date1
    if y2 < y1:
        return date2
    if m1 < m2:
        return date1
    if m2 < m1:
        return date2
    if d2 < d1:
        return date2
    return date1


def daysInBetween(date1: (int), date2: (int)) -> int:
    
    if returnEarlierDate(date1, date2) == date2:
        date1, date2 = date2, date1
        
    d1, m1, y1 = date1
    d2, m2, y2 = date2
    
    
    if y1 != y2:
    
        noOfDays = DAYS_IN_MONTHS[m1] - d1
        updateFebruary(y1)
        
        for month in range(m1 + 1, 13):
            noOfDays += DAYS_IN_MONTHS[month]
            
        for year in range(y1 + 1, y2):
            
            if isLeapYear(year):
                noOfDays += 366
            else:
                noOfDays += 365
                
        
        updateFebruary(y2)
        for month in range(1, m2):
            noOfDays += DAYS_IN_MONTHS[month]
            
        noOfDays += d2
        return noOfDays
    
    if m1 != m2:
        
        updateFebruary(y1)
        noOfDays =  DAYS_IN_MONTHS[m1] - d1 + d2
    
        for month in range(m1 + 1, m2):
            noOfDays += DAYS_IN_MONTHS[month]
            
        return noOfDays
    
    return d2 - d1
        
date1 = createRandomDate()
date2 = createRandomDate()
print(date1)
print(date2)
print(daysInBetween(date1, date2))
```

``14. Write a program to carry out the following operations on the given set
    S = {12, -2, 56, 1, 67, 3}``  
    - number of items in set S  
    - maximum element in set S    
    - minimum element in set S    
    - sum of all element in set S    
    - obtain a new sorted set from S, set S remaining unchanged   
    - report whether 40 is a member of the set    
    - report whether 1 is an element of set S      


```python
 class SetOperations:
    
    def __init__(self, set_arr: [int]):        
        self.set_arr = set_arr
        
    def getSet(self) -> [int]:
        return self.set_arr
        
    def numberOfItems(self) -> int:
        return sum([1 for each in self.set_arr])
    
    def maximumElement(self) -> int:
        maximum = self.set_arr[0]
        for each in self.set_arr:
            if each > maximum:
                maximum = each
                
        return maximum
    
    def minimumElement(self) -> int:
        minimum = self.set_arr[0]
        for each in self.set_arr:
            if each < minimum:
                minimum = each
                
        return minimum
    
    def sumOfElements(self) -> int:
        
        summation = 0
        for each in self.set_arr:
            summation = summation + each
            
        return summation
    
    def sortElements(self) -> [int]:
        
        sort_arr = [each for each in self.set_arr]
        size = self.numberOfItems()
        
        for i in range(0, size):
            for j in range(i + 1, size):
                if sort_arr[i] > sort_arr[j]:
                    sort_arr[i], sort_arr[j] = sort_arr[j], sort_arr[i]
                    
        return sort_arr
                
            
    
    def isItemPresent(self, item: int) -> bool:
        
        for each in self.set_arr:
            if each == item:
                return True
            
        return False
    
    
set_object = SetOperations([12, -2, 56, 1, 67, 3])
print("S = ", set_object.getSet())
print("Number of items in set S", set_object.numberOfItems())
print("Maximum Element in set S", set_object.maximumElement())
print("Minimum Element in set S", set_object.minimumElement())
print("Sum of all elements in set S", set_object.sumOfElements())
print("New Sorted set", set_object.sortElements())
print("Set S = ", set_object.getSet())

if (set_object.isItemPresent(40)):
    print("40 is present in S")
else:
    print("40 is not present in S")
    
if (set_object.isItemPresent(1)):
    print("1 is present in S")
else:
    print("1 is not present in S")
```

``16. Write a program to sort a dictionary in ascending / descending order by value.``

```python
import sys
def sortDictionaryByValue(dictionary: {int, int}) -> {int, int}:
    
    sorted_dict = {}
    
    while len(dictionary) != 0:
        
        value = sys.maxsize
        key = 0
        
        for each in dictionary:
            
            if dictionary[each] < value:
                value = dictionary[each]
                key = each
                
        del dictionary[key]
        sorted_dict[key] = value
        
    return sorted_dict

n = int(input())
dictionary = {}

for i in range(n):
    key, value = list(map(int, input().split(":")))
    dictionary[key] = value
    
sorted_dictionary = sortDictionaryByValue(dictionary)
print(sorted_dictionary)
```
